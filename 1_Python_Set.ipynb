# Python - Set

Python list is a collection of unordered, mutable, hetrogenous, no duplicate elements. 

Syntax for creating a set - set() / {}

```python
# defining a set using set() constructor

set_A = set((1,2,3,4,5))
print("set_A:", set_A)

set_B = set((3,4,5,6,7))
print("set_B:", set_B)
```

```python
# adding an element in a set

set_A.add(0)

print("added an element in a set:", set_A)
```

```python
# combining set_A and set_B without updating the values of the main set

set_union = set_A.union(set_B)

print("combining values of the two sets:", set_union)
```

```python
# combining set_A and set_B with updating the values of the main set

set_A.update(set_B)

print("updating the set_A:", set_union)
```

```python
# fetch all elements of set_A except which are not intersecting with set_B

set_A = set((1,2,3,4,5))

set_diff = set_A.difference(set_B)

print("all elements of the set_A but not from set_B:", set_diff)
```

```python
# fetch all elements of set_A except which are not intersecting with set_B and then updating the new values in set_A

set_A = set((1,2,3,4,5))

set_A.difference_update(set_B)

print("all elements of the set_A but not from set_B and then updating the new values in set_A:", set_A)
```

```python
# fetch common elements of set_A and set_B

set_A = set((1,2,3,4,5))

set_intersection = set_A.intersection(set_B)

print("common elements of set_A and set_B:", set_intersection)
```

```python
# fetch common elements of set_A and set_B with replacing those common elements in set_A

set_A = set((1,2,3,4,5))

set_A.intersection_update(set_B)

print("common elements of set_A and set_B and then updating the new values in set_A:", set_A)
```

```python
# fetch uncommon elements of set_A and set_B

set_A = set((1,2,3,4,5))

set_sym_diff = set_A.symmetric_difference(set_B)

print("Uncommon elements of the set_A and set_B:", set_sym_diff)
```

```python
# fetch uncommon elements of set_A and set_B and then updating the new values in set_A

set_A = set((1,2,3,4,5))

set_A.symmetric_difference_update(set_B)

print("Uncommon elements of the set_A and set_B and then updating the new values in set_A:", set_A)
```

```python
# remove a specific element from a set

set_A = set((1,2,3,4,5))

set_A = set_A.discard(5)

print("Remove an element from the set_A:", set_A)
```

```python
# remove any arbitary element from a list 

set_A = set((1,2,3,4,5))

print("is set_A superset of set_B:", set_A.pop())
```

```python
# set_A and set_B are mutually exlusive

set_A = set((1,2,3,4,5))

print("are the set_A and set_B mutually exlcusive:", set_A.isdisjoint(set_B))
```

```python
# is set_A subset of set_B

set_A = set((1,2,3,4,5))

print("is set_A a subset of set_B:", set_A.issubset(set_B))
```

```python
# is set_A superset of set_B

set_A = set((1,2,3,4,5))

print("is set_A a superset of set_B:", set_A.issuperset(set_B))
```

```python
# remove all elements from the set

set_A = set((1,2,3,4,5))

set_A.clear()

print("remove all elements of set_A:", set_A)
```
