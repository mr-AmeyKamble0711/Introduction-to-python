# Python - List

Python list is a collection of ordered, mutabl, hetrogenous elements. The elements of list are called items.

Syntax for creating a list - list() / []

```python
# defining a list using the list() constructor

mylist = list((1,2,3,4,6,5,7))
print("list name:", mylist)
```

# List operation

```python
# Adding an element to the end of the list

mylist.append(8)

print("Appending an element in the list:", mylist)
```

```python
# Adding an element to any position of the list

mylist.insert(0, 10)

print("inserting an element in the list:", mylist)
```

```python
# Adding multiple elements to the end of the list

mylist.extend([11,12,13])

print("adding multiple elements in the list:", mylist)
```

```python
# Sorting the elements of the list in ascending order

mylist.sort()

print("sorting:", mylist)
```

```python
# Sorting the elements of the list in descending order

mylist.sort(reverse=True)

print("reverse sorting:", mylist)
```

```python
# frequency of an element in a list

print("frequency of an element:", mylist.count(1))
```

```python
# Number of elements in the list

print("list length:", len(mylist))
```

```python
# removing an element by giving an extact value

mylist.remove(1)

print("removing an element from the list:", mylist)
```

```python
# removing an element by giving a extact index value

print("remving an element by giving exact index value:", mylist.pop(10))
```

```python
# gives the index of an element. If there are multiple occurance of an element then it gives the index of the first occurance.

print("Index of an element:", mylist.index(11))
```

```python
# gives the index of an element. If there are multiple occurance of an element then it gives the index of the first occurance.

print("maximum value in the list:", max(mylist))
print("minimum value in the list:", min(mylist))
```

```python
# Concept of list comprehensive -  creating a new list using an existing list

mylist_even_numbers = [i for i in mylist if i%2 == 0]

print('list comprehensive:',mylist_even_numbers)
```
